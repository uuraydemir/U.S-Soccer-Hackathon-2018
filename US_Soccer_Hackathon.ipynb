```python
# reading data
wc_data = pd.read_csv('Desktop/Hackathon Raw Files/Full Datasets - Opta/WC/World Cup.csv')
wc_data.head()
```

```python
# filtering columns
clean_df = pd.DataFrame({
    "Season":wc_data['season'],\
    'team' : wc_data['team'],\
    'team_id': wc_data['team_id']
})
clean_df.head()
```

```python
# Number of unique teams for each world cup
count_qualify_team = clean_df.groupby('team').Season.nunique()
count_qualify_team.head()
```

```python
#dataframe for occurences of teams for 4 world cup
count_qualify_team_df = pd.DataFrame({
    'Count': count_qualify_team
})
count_qualify_team_df.head()
```

```python
#filtering out more than 3 occurences
count_qual_4_df = count_qualify_team_df.loc[count_qualify_team_df['Count'] > 3]
count_qual_4_df.reset_index()
```

```python
#Number of unique games for each season and team
count_team = wc_data.groupby(['season','team']).game_id.nunique()
count_team_df = pd.DataFrame({
    'Count' : count_team
})
count_team_df.to_csv('Team_1stFilter.csv')
```

```python
#dataframe show
filtered_1st_data = pd.read_csv('Team_1stFilter.csv')
filtered_1st_data.head()
```

```python
#Qualifying 4 times to World Cup & more than 4 matches (more than 4 matches means that at least quarter final) 
filtered_teams_ = filtered_1st_data[(filtered_1st_data['Count'] > 4) & (filtered_1st_data['team'].isin(["Argentina", 'Australia', 'Brazil', 'England', 'France', 'Germany', 'Japan', 'Korea Republic', 'Mexico', 'Portugal', 'Spain', 'Switzerland']))]
top_teams_array = filtered_teams_.team.unique()
top_teams_array
```

```python
#if event_type == 1 , then it is pass
top_team_all_data = wc_data.loc[wc_data['team'].isin(top_teams_array)]
top_teams_pass = top_team_all_data.loc[top_team_all_data['event_type_id'] == 1]
top_teams_pass.head()
```

```python
#total passes 
teams_pass_count = top_teams_pass.groupby('team').event_type_id.count()
pd.to_numeric(teams_pass_count)
```

```python
# successful passes
top_teams_pass_out = top_teams_pass.loc[top_teams_pass['outcome'] == 1]
success_pass = top_teams_pass_out.groupby('team').outcome.count()
pd.to_numeric(success_pass)
```

TOP TEAMS PASS ACCURACY

```python
#pass accuracy df
pass_accuracy_df = pd.DataFrame({
    'Successful Pass':success_pass,
    'Total Pass Attempt': teams_pass_count,
    'Pass Accuracy': success_pass/teams_pass_count
    })
pass_accuracy = pass_accuracy_df.reset_index()
pass_accuracy
```

US SOCCER NATIONAL TEAM PASS DATA

```python
#us all data read
us_all_data = pd.read_csv('Hackathon Raw Files/Full Datasets - Opta/USMNT/USMNT.csv')
us_all_data.head()
```

```python
#us pass data
us_pass = us_all_data.loc[(us_all_data['event_type_id'] == 1) & (us_all_data['team'] == 'USA') ]
us_pass.head()
```

```python
#us total passes count
us_pass_count = us_pass.groupby('team').event_type_id.count()
pd.to_numeric(us_pass_count)
us_pass_count
```

```python
# us successful passes
us_teams_pass_out = us_pass.loc[us_pass['outcome'] == 1]
us_success_pass = us_teams_pass_out.groupby('team').outcome.count()
pd.to_numeric(us_success_pass)
us_success_pass
```

USA PASS ACCURACY

```python
#US pass accuracy df
us_pass_accuracy_df = pd.DataFrame({
    'Successful Pass':us_success_pass,
    'Total Pass Attempt': us_pass_count,
    'Pass Accuracy': us_success_pass/us_pass_count
    })
us_pass_accuracy = us_pass_accuracy_df.reset_index()
us_pass_accuracy
```
